```python
import pandas as pd
import zipfile
import io
import requests

# URL of the ZIP file on GitHub
url = "https://github.com/KungumapriyaaMKP/InfosysSpringboard/raw/refs/heads/main/archive%20(4).zip"

# Download the ZIP
response = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(response.content))

# List files inside the ZIP (optional)
print("Files in ZIP:", z.namelist())

# Adjust the filename below if it's different
csv_filename = "synthetic_text_cleaning_realworld.csv"

# Read the CSV into a pandas DataFrame
with z.open(csv_filename) as f:
    df = pd.read_csv(f)

print("Loaded dataset with shape:", df.shape)
df.head()
```

```python
import re
import string

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = str(text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\+?\d[\d \-\(\)]{7,}\d', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[@#]\w+', '', text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    return text.lower()

# cleaning all text/object columns
text_cols = df.select_dtypes(include='object').columns
df[text_cols] = df[text_cols].applymap(clean_text)

df.head()
```

```python
# Save as Excel
cleaned_filename = "cleaned_dataset.xlsx"
df.to_excel(cleaned_filename, index=False)

from google.colab import files
files.download(cleaned_filename)
```
